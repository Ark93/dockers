### Running docker image

```python
# creating container
container_id = client.create_container(image='jupyter/scipy-notebook', 
                                       command='ls',
                                       ports=[8888],
                                       volumes='/mnt/vol1',
                                       host_config=client.create_host_config(
                                           port_bindings={8888:10000},
                                           binds={
                                               '/home/santos/Documents/Repos': {
                                                   'bind': '/mnt/vol1',
                                                   'mode': 'rw',
                                               }
                                           }
                                       ),         
                                  )
```

```python
print(container_id)
```

```python
!docker run --name=jupyter -it -p 10000:8888 -e JUPYTER_ENABLE_LAB=yes -v /home/santos/Documents/Repos:/home/jovyan/Repos jupyter/scipy-notebook:latest 
```

explanation of previous command

1. docker run : command for running and creating containers

2. --name=jupyter : name container

3. -it : command indicates the use of terminal in created container

4. -p 10000:8888  : port binding of host:container. Jupyter notebook run in port 8888 and host can access thru port 10000

5. -e JUPYTER_ENABLE_LAB=yes : indicates to jupyter container to enable Jupyter Lab

6. -v  /home/santos/Documents/Repos:/home/jovyan/Repos : bindmount indicated folder from host system to container system  'host:container'

7. jupyter/scipy-notebook:latest  : image to run
